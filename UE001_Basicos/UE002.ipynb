


### Workloads esenciales para Unreal Engine & C++

Para trabajar con Unreal Engine de manera efectiva, asegúrate de tener seleccionados los siguientes apartados:

1.  **Desarrollo de juegos con C++ (Fundamental):**
    *   Este es el componente **imprescindible**. Incluye el compilador de C++, las herramientas de depuración y las bibliotecas necesarias para que Unreal Engine pueda compilar el código de tu proyecto.
    *   *Nota:* Dentro de los detalles de instalación (a la derecha), asegúrate de que esté marcada la opción **"Compatibilidad con el IDE de Unreal Engine"**.

2.  **Desarrollo para el escritorio con C++:**
    *   Aunque Unreal tiene su propio flujo, este workload instala herramientas base de Windows (como el SDK de Windows) que son necesarias para compilar cualquier aplicación de escritorio en este sistema operativo.

### Otros componentes en tu imagen
*   **Desarrollo de juegos con Unity:** No es necesario para Unreal. Puedes desmarcarlo si quieres ahorrar espacio en disco.
*   **Desarrollo de escritorio de .NET / WinUI:** No son requeridos para Unreal Engine, a menos que planees crear herramientas externas adicionales para tu juego usando esas tecnologías.

---

**Sugerencia:** Revisa en la columna de la derecha (**Detalles de la instalación**) que el **SDK de Windows** más reciente esté marcado, ya que Unreal lo requiere para vincular los binarios correctamente.





### Selección de componentes para Unreal y Meta Quest 3

Basado en la lista de "Desarrollo para el escritorio con C++" que muestras:

#### (a) Para Unreal Engine (Base)
*   **Herramientas de compilación de C++ de MSVC v143:** (Es la primera opción bajo 'Opcional'). Es el compilador principal.
*   **SDK de Windows 11 (10.0.22621.0):** Indispensable para que el motor funcione en Windows.
*   **IntelliCode:** Muy útil para el autocompletado de funciones de Unreal.

#### (b) Para Meta Quest 3 (AR/Android)
*   **Herramientas de Clang de C++ para Windows:** (Aparece desmarcada en tu imagen). **Márcala**. Unreal la utiliza para la compilación cruzada hacia Android (sistema operativo de las Quest).

#### (c) ¿Qué NO es importante? (Puedes desmarcarlo para ahorrar espacio)
*   **ATL de C++:** No se usa para el desarrollo estándar de juegos.
*   **C++ Build Insights:** Es para optimizar tiempos de compilación en proyectos masivos; no lo necesitas ahora.
*   **Adaptador de prueba para Boost.Test y Google Test:** Unreal tiene su propio sistema de pruebas; estos solo ocupan espacio.
*   **C++ MFC:** Tecnología antigua para ventanas de Windows, totalmente innecesaria para Unreal.
*   **MSVC v142 / v141:** Son versiones viejas (VS 2019/2017). Si usas Unreal Engine 5, quédate solo con la **v143**.

Basado en tu nueva imagen de "Desarrollo de juegos con C++":

#### (a) Para Unreal Engine (General)
*   **Visual Studio Tools for Unreal Engine:** ¡CRÍTICO! Es lo que permite que Visual Studio reconozca las macros y clases de Unreal.
*   **herramientas del depurador de Visual Studio para Unreal...:** Indispensable para encontrar errores (bugs) mientras el motor está ejecutándose.
*   **Herramientas HLSL:** Muy importante si planeas crear tus propios materiales o efectos visuales (shaders).

#### (b) Para Meta Quest 3 (AR / VR / Android)
*   **Android Visual Studio Tools for Unreal Engine:** Este es el componente **más importante** para las Quest 3 en esta lista. Las gafas de Meta usan Android, y esta herramienta permite depurar y desplegar el código directamente al dispositivo.

#### (c) ¿Qué NO es importante? (Puedes desmarcarlo)
*   **Cocos:** Es un motor de juegos totalmente diferente. No tiene nada que ver con Unreal.
*   **administrador de paquetes vcpkg:** Unreal ya trae sus propias librerías; no necesitas este gestor externo para empezar.
*   **C++ Build Insights:** Es una herramienta avanzada para medir cuánto tarda en compilar tu código. Útil en empresas grandes, innecesario para empezar.
*   **Incredibuild:** Solo funciona si tienes una licencia de pago costosa. Si no la tienes, solo ocupa espacio.

### Componentes Individuales y Extensiones

De lo que se ve en esta captura, esto es lo que debes saber:

#### 1. Importantes para Unreal Engine (C++)
*   **Herramientas de compilación de MSVC v143...**: ¡Aquí está la que buscábamos! Esta es la versión más reciente del compilador para Visual Studio 2022. **Mantenla marcada**, es el motor que traduce tu código a algo que la computadora entiende.
*   **Herramientas de compilación de C++ de MS...**: (Es la que está justo encima). Probablemente sea el set de herramientas base. Déjala marcada para asegurar compatibilidad.

#### 2. Útil para Gráficos y Shaders
*   **GLSL language integration (Extensiones)**: Es útil si vas a escribir código de sombreadores (shaders) personalizados. Aunque Unreal usa principalmente HLSL, tener soporte para GLSL no estorba y ayuda si importas recursos que lo usen.

#### 3. No son importantes (Opcionales)
*   **Live Share**: Solo es útil si vas a programar en tiempo real con otra persona de forma remota en el mismo archivo. Si trabajas solo, puedes quitarlo.
*   **Paquete de compatibilidad de .NET Framework...**: Unreal Engine 5 usa principalmente C++, pero algunas herramientas internas del editor usan .NET. Normalmente el instalador de Unreal ya trae lo que necesita, por lo que este paquete es opcional aquí.
