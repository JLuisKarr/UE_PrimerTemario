# Selección de Clase Padre (Parent Class)

Estás en el paso más importante para definir qué **"naturaleza"** tendrá tu objeto en el código. En Unreal, todo funciona por **jerarquías** (herencia): eliges una base que ya sabe hacer cosas y tú le añades lo nuevo.

### ¿Qué significan estas opciones?

*   **None (Ninguna):** Es una clase de C++ pura y vacía. No tiene "cuerpo" en el mundo 3D. Se usa para cálculos matemáticos o gestión de datos que no necesitan verse en pantalla.
*   **Actor (La que tienes seleccionada):** Es la base para cualquier cosa que "exista" en el nivel. 
    *   **¿Por qué elegirla?** Si quieres crear un cubo, un cofre o una moneda, esta es la correcta. Un Actor puede tener posición, rotación y escala.
*   **Pawn (Peón):** Es un Actor, pero con una habilidad extra: puede ser "poseído" por un jugador o por una Inteligencia Artificial (IA) para recibir órdenes de movimiento.
*   **Character (Personaje):** Es un Pawn especializado en seres humanos o criaturas bípedas. Ya incluye de fábrica la lógica para caminar, saltar y caer (gravedad).

### Resumen de tu elección actual: **Actor**
Al elegir **Actor** y darle a "Next", Unreal creará dos archivos para tu cubo:
1.  **`.h` (Header):** Donde declaras "qué tiene" el cubo (su nombre, su velocidad, su color).
2.  **`.cpp` (Source):** Donde escribes "qué hace" el cubo (moverse, girar, explotar).

Ahora estás viendo el archivo **Header (.h)**. Si el archivo `.cpp` es el "músculo", este es el **"plano"**. Aquí es donde debes declarar qué componentes tendrá tu cubo (como su malla o su forma) para que el motor sepa que existen.

### Lo que sigue para que el cubo sea visible:

Para que tu Actor tenga forma de cubo, necesitas declararlo aquí. Sigue estos pasos:

1.  **Añadir el componente:** Debajo de la línea 15 (`AMiCubo();`), crea un espacio y pega este código:

```cpp
public:
	// Esto crea el espacio para la malla del cubo
	UPROPERTY(VisibleAnywhere)
	UStaticMeshComponent* MallaCubo;
```

### ¿Qué significa este código?
*   **UStaticMeshComponent:** Es el tipo de objeto que Unreal usa para mostrar modelos 3D (como cubos, esferas o personajes).
*   **UPROPERTY(VisibleAnywhere):** Esta es una "macro" de Unreal. Le dice al editor: "Haz que esta variable sea visible en el panel de detalles para que pueda verla y cambiar el modelo del cubo desde Unreal".

# Archivo MiCubo.cpp
```cpp
#include "MiCubo.h"

// Constructor: Aquí "armamos" el objeto
AMiCubo::AMiCubo()
{
    // 1. Permitir que el objeto se actualice cada frame
    PrimaryActorTick.bCanEverTick = true;

    // 2. Crear físicamente el componente de malla
    MallaCubo = CreateDefaultSubobject<UStaticMeshComponent>(TEXT("MallaDelCubo"));

    // 3. Hacer que la malla sea la raíz del objeto (el centro)
    RootComponent = MallaCubo;
}

// Se ejecuta al dar Play
void AMiCubo::BeginPlay()
{
	Super::BeginPlay();
}

// Se ejecuta cada frame
void AMiCubo::Tick(float DeltaTime)
{
	Super::Tick(DeltaTime);

    // OPCIONAL: Descomenta la línea de abajo para que el cubo rote solo
    // AddActorLocalRotation(FRotator(0.f, 50.f * DeltaTime, 0.f));
}
```

# Archivo MiCubo.h
```cpp
#pragma once

#include "CoreMinimal.h"
#include "GameFramework/Actor.h"
#include "MiCubo.generated.h"

UCLASS()
class CUBOBASICO02_API AMiCubo : public AActor
{
	GENERATED_BODY()
	
public:	
	// Constructor
	AMiCubo();

protected:
	// Se llama cuando el juego comienza
	virtual void BeginPlay() override;

public:	
	// Se llama en cada fotograma
	virtual void Tick(float DeltaTime) override;

	// --- ESTO ES LO QUE HEMOS AÑADIDO ---
	// UPROPERTY permite que Unreal vea esta variable en el editor
	// VisibleAnywhere significa que podremos ver sus propiedades en el panel de Detalles
	UPROPERTY(VisibleAnywhere, Category = "Mi Contenido")
	class UStaticMeshComponent* MallaCubo;
};
```

# Bitácora de Resolución: Creación de Actor C++ en Unreal Engine 5

A continuación, presento el resumen técnico de los obstáculos encontrados durante la creación del proyecto **cuboBasico02** y las soluciones aplicadas para alcanzar un estado funcional.

---

### 1. Error de Compilación Inicial (Entorno)
*   **Problema:** Al abrir el proyecto, Visual Studio reportaba errores en scripts de automatización (`CA2017`, `CS8604`) y fallos en paquetes NuGet.
*   **Solución:** Se identificó que estos errores eran internos de la versión del motor y el SDK de .NET. La solución fue **ignorar la compilación desde Visual Studio** y delegar la tarea al sistema de **Live Coding** de Unreal, que es más permisivo con el estado del editor.

### 2. Conflicto de Escritura (Live Coding vs Visual Studio)
*   **Problema:** Error `Unable to build while Live Coding is active`. Visual Studio no podía compilar porque el editor de Unreal tenía los archivos bloqueados.
*   **Solución:** Se estableció un flujo de trabajo binario: **escribir** en Visual Studio y **compilar** exclusivamente desde el editor de Unreal usando el atajo `Ctrl + Alt + F11`.

### 3. Clase Invisible (Lógica de C++)
*   **Problema:** Al arrastrar la clase `MiCubo` al nivel, no se veía nada (solo un punto en el espacio).
*   **Solución:** Se modificaron los archivos fuente para integrar un componente visual:
    *   **En `.h`:** Se declaró un puntero a `UStaticMeshComponent` con la macro `UPROPERTY(VisibleAnywhere)`.
    *   **En `.cpp`:** Se utilizó `CreateDefaultSubobject` en el constructor para instanciar físicamente la malla y se asignó como `RootComponent`.

### 4. Error de Sintaxis C++ (El Punto y Coma)
*   **Problema:** El archivo `.h` marcaba error rojo en la última línea y en las macros de Unreal.
*   **Solución:** Se detectó la falta de un punto y coma (`;`) al cerrar la llave de la clase `AMiCubo`. Tras añadirlo, se sincronizó el archivo para que el motor pudiera procesar las macros correctamente.

### 5. Error de Linkeo de JSON (Fallo Crítico de Live Coding)
*   **Problema:** La consola de Live Coding mostró el error `missing LinkerPath field` en `LiveCoding.json`, impidiendo cualquier actualización de código.
*   **Solución:** Se realizó una limpieza profunda:
    1. Cierre total de programas.
    2. Borrado manual de las carpetas `Binaries` e `Intermediate`.
    3. Regeneración de archivos de proyecto (`.uproject` -> Generate Visual Studio project files).
    4. Reconstrucción completa (Rebuild) al abrir el motor.

### 6. Visibilidad y Física en el Viewport
*   **Problema:** El cubo no caía al dar Play o se perdía detrás de la cámara.
*   **Solución:** 
    *   **Visual:** Se asignó manualmente el *Static Mesh* "Cube" en el panel de Detalles.
    *   **Física:** Se activó la casilla **Simulate Physics** para habilitar la gravedad.
    *   **Cámara:** Se ajustaron las coordenadas de *Location* a (0,0,200) para centrar el objeto frente a la cámara inicial.

---

**Estado Final:** El proyecto cuenta con un Actor programado en C++ que hereda propiedades físicas y es capaz de ejecutar lógica en tiempo real (Tick).
