

# Guía de Recuperación: Del "Crash" al Proyecto Estable

Este proceso se activó tras el bloqueo de Unreal Engine y la corrupción del archivo del mapa.

### 1. Limpieza de Archivos Corruptos
Cuando Unreal se congela y no abre, es porque los archivos temporales están dañados.
*   **Acción:** Cerramos el motor y borramos manualmente las carpetas `Binaries`, `Intermediate` y `Saved`.
*   **Resultado:** Eliminamos cualquier rastro de la compilación fallida que bloqueaba el inicio.

### 2. Regeneración de la Solución (.uproject)
Sin la carpeta `C++ Classes`, el motor no reconoce nuestro código.
*   **Acción:** Clic derecho en el archivo `geometrias.uproject` > **"Generate Visual Studio project files"**.
*   **Resultado:** Unreal volvió a crear los archivos necesarios para conectar el código C++ con el editor.

### 3. El "Truco del Click Derecho" en la Solución
Este paso fue vital para ignorar errores internos de los plugins de Epic y centrarnos en nuestro código.
*   **Acción:** Abrimos Visual Studio y, en el Explorador de Soluciones, buscamos la carpeta **Games** > **geometrias**.
*   **Acción:** Hicimos **Clic Derecho** solo sobre nuestro proyecto (`geometrias`) y seleccionamos **Compilar (Build)**.
*   **Configuración:** Cambiamos el modo a **Development Editor** en la barra superior.
*   **Resultado:** Logramos el mensaje `1 correcto, 0 erróneo`, permitiendo que Unreal abriera sin mensajes de "Missing Modules".

### 4. Reparación de Actores en el Mapa
Tras el cierre forzado, el nivel detectó "archivos inválidos".
*   **Acción:** Usamos la herramienta **"Repair Errors"** en el Outliner.
*   **Resultado:** Limpiamos las referencias fantasmas de la instancia de `MiGeometria` que se había corrompido.

### 5. Configuración de Maps & Modes (Persistencia)
Para que el trabajo no se pierda al cerrar, configuramos los mapas de inicio.
*   **Paso A:** Guardamos el mapa manualmente como `MapaPrincipal`.
*   **Paso B:** En **Project Settings > Maps & Modes**, asignamos `MapaPrincipal` en:
    *   *Editor Startup Map* (Para abrirlo al trabajar).
    *   *Game Default Map* (Para cuando el juego se ejecute).

### 6. Restauración de la Interfaz (Layout)
Recuperamos las herramientas de trabajo que habían desaparecido.
*   **Acción:** Activamos desde el menú **Window** los paneles **Outliner** y **Details**.
*   **Acción:** Los anclamos a la derecha y guardamos el diseño en **Window > Save Layout**.

### 7. Verificación Final (El Asterisco)
*   **Check:** Confirmamos que en la pestaña superior de `MapaPrincipal` **no hubiera un asterisco (*)**.
*   **Check:** El icono de **"All Saved"** en verde confirmó que el disco duro tiene la versión final.
