


# Interfaz de Unreal Project Browser

La imagen muestra el **Unreal Project Browser**, que es la ventana de inicio de **Unreal Engine 5** donde los usuarios configuran y crean nuevos proyectos.

### Secciones Principales:

*   **Panel Izquierdo (Categorías):** Permite elegir el flujo de trabajo según la industria:
    *   **Games:** Seleccionada actualmente; ofrece plantillas para mecánicas de juego.
    *   **Film / Video & Live Events:** Para producción virtual y cinemáticas.
    *   **Architecture:** Para visualización arquitectónica e interactividad.
    *   **Automotive, Product Design & Manufacturing:** Para renders de alta fidelidad de productos y vehículos.
    *   **Simulation:** Para simulaciones técnicas o de entrenamiento.

*   **Panel Central (Plantillas):** Muestra bases preconfiguradas para juegos:
    *   **Blank:** Un proyecto vacío (seleccionado).
    *   **First Person / Third Person:** Cámaras en primera o tercera persona.
    *   **Vehicle:** Física de vehículos.
    *   **Virtual Reality / Handheld AR:** Para dispositivos de realidad aumentada y virtual.

*   **Panel Derecho (Configuración del Proyecto):**
    *   **Blueprint vs C++:** El usuario ha elegido **Blueprint**, el sistema de programación visual por nodos.
    *   **Target Platform:** Configurado para **Desktop** (PC/Consola).
    *   **Quality Preset:** Ajustado en **Maximum** para mejores gráficos.
    *   **Starter Content:** Casilla marcada para incluir materiales y sonidos básicos.
    *   **Raytracing:** Casilla desmarcada (iluminación por trazado de rayos desactivada).

*   **Barra Inferior:**
    *   **Project Location:** El proyecto se guardará en `D:\Unreal\Basicos`.
    *   **Project Name:** El nombre asignado es `MyProject`.




# Guía: Creación de un Proyecto en Blanco en Unreal Engine

Basado en la configuración que tienes seleccionada en la imagen, aquí tienes el desglose de lo que significa crear un proyecto bajo estos parámetros:

### 1. Selección de Plantilla: "Blank"
Al elegir **Blank** (en azul), estás optando por un lienzo totalmente vacío. 
*   **Ventaja:** No hay código ni activos innecesarios que limpiar; es ideal para aprender desde cero o para proyectos con mecánicas muy personalizadas.
*   **Desventaja:** A diferencia de las plantillas *Third Person* o *First Person*, aquí no tendrás un personaje que se mueva ni una cámara configurada por defecto.

### 2. Configuración Técnica (Project Defaults)
*   **Blueprint:** Has seleccionado el sistema de programación visual. Esto te permitirá crear la lógica de tu "cubo" conectando nodos sin necesidad de escribir código complejo en C++.
*   **Target Platform: Desktop:** El motor optimizará los recursos para ser ejecutados en computadoras o consolas de sobremesa.
*   **Quality Preset: Scalable:** A diferencia de la imagen anterior, aquí has seleccionado "Scalable". Esto significa que el proyecto tendrá una configuración gráfica flexible que puede ajustarse automáticamente para funcionar en hardware menos potente.
*   **Starter Content (Activado):** Es una excelente decisión para un proyecto en blanco. Incluye una carpeta con materiales, sonidos y mallas básicas (como esferas y cubos) que te servirán para empezar a prototipar de inmediato.
*   **Raytracing (Desactivado):** Esto mejora el rendimiento mientras trabajas, ya que no calcula trazado de rayos en tiempo real, algo que suele reservarse para la etapa final de renderizado o equipos de gama muy alta.

### 3. Identificación y Ruta
*   **Nombre:** El proyecto se llamará `cuboBasico02`.
*   **Ubicación:** Se guardará en el disco `D:`, dentro de tu carpeta de estudios `Unreal\Basicos`.
