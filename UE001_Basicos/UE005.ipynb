# Notas de Clase: Programación Orientada a Objetos en Unreal Engine

### (a) Programación Orientada a Objetos (POO)
Es un paradigma de programación basado en el concepto de "objetos", que pueden contener datos (atributos) y código (métodos). En lugar de ver el programa como una lista de instrucciones, se ve como un conjunto de piezas que interactúan entre sí.

### (b) Clase en POO y su uso en UE
*   **Concepto:** Es el "plano" o molde. Define qué datos y qué acciones tendrá un objeto.
*   **En Unreal:** Casi todo lo que creas nace de una clase. Por ejemplo, la clase `Actor` define que algo puede estar en el nivel, mientras que `Character` añade la capacidad de caminar.

### (c) Instancia en POO y su uso en UE
*   **Concepto:** Es el objeto real creado a partir del molde (clase).
*   **En Unreal:** Si la clase es "Silla", la **instancia** es la silla específica que arrastraste al mapa. Puedes tener 10 instancias de la misma clase, cada una en una posición diferente.

### (d) Constructor de clase y su uso en UE
*   **Concepto:** Una función especial que se ejecuta automáticamente cuando se crea el objeto. Sirve para dar valores iniciales.
*   **En Unreal:** Se usa principalmente para crear componentes. 
    *   *Ejemplo:* `AMiCubo::AMiCubo() { Mesh = CreateDefaultSubobject<UStaticMeshComponent>(TEXT("Mesh")); }`

### (e) Función en programación
Es un bloque de código diseñado para realizar una tarea específica que se puede reutilizar varias veces en el programa sin tener que volver a escribirlo.

### (f) PUBLIC, PROTECTED y PRIVATE
Definen quién puede ver o usar los elementos de la clase:
*   **Public:** Accesible desde cualquier otra clase o Blueprint.
*   **Protected:** Accesible solo por la propia clase y sus "hijos" (clases derivadas).
*   **Private:** Accesible **solo** dentro de esa clase específica.

### (g) Funciones dentro de una clase en UE
Sirven para definir el comportamiento del objeto. Permiten que el objeto reaccione a eventos (ej. saltar, recibir daño, abrir una puerta) o realice cálculos internos.

### (h) UCLASS()
*   **Concepto:** Es una "macro" de Unreal que le dice al motor: "Oye, esta clase de C++ no es normal, quiero que la reconozcas y la muestres en el Editor".
*   **Ejemplo:**
    ```cpp
    UCLASS()
    class MI_PROYECTO_API AMiCubo : public AActor { ... };
    ```

### (i) Funciones Principales de Ciclo de Vida
*   **BeginPlay():** Se ejecuta una sola vez cuando el juego inicia o el objeto aparece en el mundo. Útil para configuraciones iniciales.
    *   *Ejemplo:* Desactivar la gravedad de un objeto al empezar.
*   **Tick(float DeltaTime):** Se ejecuta en cada cuadro (frame) del juego. Sirve para cosas que cambian constantemente.
    *   *Ejemplo:* Hacer que un objeto rote continuamente o siga a un jugador.

### (j) UPROPERTY()
*   **Concepto:** Macro que permite que una variable de C++ sea visible y editable desde el Editor de Unreal o Blueprints. También ayuda a Unreal a gestionar la memoria (Garbage Collection).
*   **Ejemplo:**
    ```cpp
    UPROPERTY(EditAnywhere, BlueprintReadWrite, Category = "Atributos")
    float VelocidadRotacion;
    ```
